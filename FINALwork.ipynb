```
import pandas as pd

#загрузка данных

r_cols = ['user_id', 'item_id', 'rating', 'timestamp']
ratings = pd.read_csv('u_data.csv', sep='\t', names = r_cols, encoding='latin-1')

genre = pd.read_csv('u_genre.csv', sep='|', names = ('genre','id'), encoding='latin-1')

info = pd.read_csv('u_info.csv', sep=' ', names = ('quan','obj'), encoding='latin-1')

i_cols = ['movie_id', 'movie_title' ,'release_date','video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
item = pd.read_csv('u_item.csv', sep='|', names = i_cols, encoding='latin-1')

o_cols = ['occup']
occ = pd.read_csv('u_occupation.csv', names = o_cols, encoding='latin-1')

u_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_csv('u.user.csv', sep='|', names = u_cols, encoding='latin-1')
```

```
#Посчитайте среднюю оценку, которую поставил каждый пользователь

rat = ratings.filter(items = ['user_id', 'rating'])
avg_rat = rat.groupby(['user_id']).mean()
avg_rat
```

```
#Отсортируйте фильмы по количеству оценок

filtritem = item.filter(items = ['movie_id', 'movie_title'])
filtrat = ratings.filter(items = ['item_id', 'rating'])

cntfilm = filtrat.groupby(['item_id']).count().sort_values(by=['rating'], ascending=False)

film = cntfilm.join(filtritem.set_index(['movie_id']))
film
```

```
#Средствами pandasa соедините данные по оценкам с данными по пользователям (u.user)
joinuser = users.join(avg_rat, 'user_id', how='inner')
joinuser
```

```
#Построить модель, которая предсказывает оценку для фильма и пользователя 
#(воспользоваться RandomForestRegressor, тк мы решаем задачу регрессии)

data = ratings.filter(items = ['user_id', 'item_id', 'rating'])

x = data[['item_id', 'rating']]
y = data[['user_id', 'item_id']]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import *

model =  RandomForestRegressor(bootstrap=True, max_depth=30,
           max_features='auto',
           min_samples_leaf=10, min_samples_split=100,
           n_estimators=300, 
           oob_score=True, random_state=None)

model.fit(x_train,y_train) # обучение
mod_pred = model.predict(x_test) # предсказание
```

```
#подбор параметров
from pprint import pprint
import numpy as np

print('Parameters currently in use:\n')
pprint(model.get_params())

from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

pprint(random_grid)

model = RandomForestRegressor()
```

```
#Оценить качество регрессора на отложенной (train_test_split) выборке по одной из метрик на выбор - RMSE, RMAE, R^2  
r2_score = r2_score(y_test, mod_pred)
print("r^2 on test data : %f" % r2_score)
```

```
rmse = np.sqrt(mean_squared_error(y_test, mod_pred))
print("Root Mean Squared Error: {}".format(rmse))
```

```
schema1 = StructType([
    StructField("user_id", IntegerType(), True),
    StructField("item_id", IntegerType(), True),
    StructField("rating", IntegerType(), True),
    StructField("timestamp", IntegerType(), True)])
spark_rat = spark.read.csv('u_data.csv',  sep='\t', inferSchema=True, header=False, schema = schema1)
spark_rat.show(10)
```

```
schema3 = StructType([
    StructField("quan", IntegerType(), True),
    StructField("obj", StringType(), True)])
spark_info = spark.read.csv('u_info.csv',  sep=' ', inferSchema=True, header=False, schema = schema3)
spark_info.show(10)
```

```
schema4 = StructType([
    StructField("movie_id", IntegerType(), True),
    StructField("movie_title", StringType(), True),
    StructField("release_date", StringType(), True),
    StructField("video_release_date", StringType(), True),
    StructField("IMDb_URL", StringType(), True),
    StructField("unknown", IntegerType(), True),
    StructField("Action", IntegerType(), True),
    StructField("Adventure", IntegerType(), True),
    StructField("Animation", IntegerType(), True),
    StructField("Children", IntegerType(), True),
    StructField("Comedy", IntegerType(), True),
    StructField("Crime", IntegerType(), True),
    StructField("Documentary", IntegerType(), True),
    StructField("Drama", IntegerType(), True),
    StructField("Fantasy", IntegerType(), True),
    StructField("FilmNoir", IntegerType(), True),
    StructField("Horror", IntegerType(), True),
    StructField("Musical", IntegerType(), True),
    StructField("Mystery", IntegerType(), True),
    StructField("Romance", IntegerType(), True),
    StructField("SciFi", IntegerType(), True),
    StructField("Thriller", IntegerType(), True),
    StructField("War", IntegerType(), True),
    StructField("Western", IntegerType(), True)])
spark_item = spark.read.csv('u_item.csv',  sep='|', inferSchema=True, header=False, schema = schema4)
spark_item.show(10)
```

```
schema5 = StructType([
    StructField("occupation", StringType(), True)])
spark_occ = spark.read.csv('u_occupation.csv', inferSchema=True, header=False, schema = schema5)
spark_occ.show(10)
```

```
schema6 = StructType([
    StructField("user_id", IntegerType(), True),
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(), True),
    StructField("occupation", StringType(), True),
    StructField("zip_code", IntegerType(), True)])
spark_users = spark.read.csv('u.user.csv',  sep='|', inferSchema=True, header=False, schema = schema6)
spark_users.show(10)
```

```
#Средствами спарка вывести среднюю оценку для каждого фильма
spark_rat_lim = spark_rat.select('item_id', 'rating')
spark_film_lim = spark_item.select('movie_id','movie_title')

avg_sp = spark_rat_lim.groupby(['item_id']).mean('rating').orderBy('avg(rating)', ascending=False)

avg_sp.join(spark_film_lim, spark_rat_lim.item_id == spark_item.movie_id).show()

```

```
#В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами 
#по количеству оценок, либо по самой оценке - на Ваш выбор

cnt_sp_top = spark_rat_lim.groupby(['item_id']).count().orderBy('count', ascending=False)
cnt_sp_top.join(spark_film_lim, spark_rat_lim.item_id == spark_item.movie_id).show(5)
```

```
cnt_sp_bottom = spark_rat_lim.groupby(['item_id']).count().orderBy('count', ascending=True)
cnt_sp_bottom.join(spark_film_lim, spark_rat_lim.item_id == spark_item.movie_id).show(5)
```

```
#Средствами спарка соедините информацию по фильмам и жанрам (u.genre) | Дублирование строк в данном случае - ок | Интересует такая структура - id фильма | название фильма | жанр.

sp = spark_item.select('movie_id', 'movie_title','unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','FilmNoir', 'Horror', 'Musical', 'Mystery', 'Romance', 'SciFi', 'Thriller', 'War', 'Western')
sp_unpvt = sp.selectExpr("movie_id", "movie_title", "stack(19, 'unknown', unknown, 'Action', Action, 'Adventure', Adventure, 'Animation', Animation, 'Children', Children, 'Comedy', Comedy, 'Crime', Crime, 'Documentary', Documentary, 'Drama', Drama, 'Fantasy', Fantasy, 'FilmNoir', FilmNoir, 'Horror', Horror, 'Musical', Musical, 'Mystery', Mystery, 'Romance', Romance, 'SciFi', SciFi, 'Thriller', Thriller, 'War', War, 'Western', Western) as (Genr, Total)").where("Total != 0")
sp_unpvt_g = sp_unpvt.join(spark_genre, sp_unpvt.Genr == spark_genre.genre)
sp_unpvt_grt = sp_unpvt_g.join(avg_sp, avg_sp.item_id == sp_unpvt_g.movie_id)
sp_unpvt_grt.show()
```
